```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from fraud import collectAllFeaturesBaseline, collectAllFeaturesBaselineTest, merge_on

df_fraud_aggregated_test = collectAllFeaturesBaselineTest()
df_fraud_aggregated = collectAllFeaturesBaseline()
```

```python
df_fraud_aggregated.head()
```

```python
df_fraud_aggregated_test.head()
```

```python
df_fraud_aggregated_test_merged = merge_on(df_fraud_aggregated_test, df_fraud_aggregated, on="region", right_column="f_t_region_fraud_rate")
df_fraud_aggregated_test_merged = merge_on(df_fraud_aggregated_test_merged, df_fraud_aggregated, on="disrict", right_column="f_t_district_target_mean")
df_fraud_aggregated_test_merged = merge_on(df_fraud_aggregated_test_merged, df_fraud_aggregated, on="client_catg", right_column="f_t_client_catg_target_mean")
```

```python
df_fraud_aggregated_test_merged.head()
```

```python
X = df_fraud_aggregated.drop(columns=["client_id", "target"])
#X = df_fraud_aggregated.drop(columns=["client_id", "target", "region", "disrict", "client_catg"])
X.fillna(0, inplace=True)  
y = df_fraud_aggregated["target"]

# 2. Train/Test-Split
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    stratify=y,
    random_state=42
)

# 3. Feature-Scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled  = scaler.transform(X_test)

# 4. Modell-Definition und Training
model = LogisticRegression(
    class_weight='balanced',  # hier werden die Klassengewichte automatisch angepasst
    solver='liblinear',       # geeignet für kleine bis mittelgroße Datensätze
    random_state=42
)
model.fit(X_train_scaled, y_train)

# 5. Evaluation
y_pred = model.predict(X_test_scaled)

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(
    estimator=model,
    X=scaler.fit_transform(X), 
    y=y,
    cv=5,
    scoring="roc_auc"   # oder "accuracy", "f1", ...
)
print("CV AUC-Scores:", scores)
print("Mean AUC:", scores.mean())
```

```python
feature_cols = X_train.columns

X_test2 = df_fraud_aggregated_test_merged[feature_cols].fillna(0)

X_test2_scaled = scaler.transform(X_test2)


y_pred_test = model.predict(X_test2_scaled)
y_proba_test = model.predict_proba(X_test2_scaled)[:, 1]  # falls du Wahrscheinlichkeiten brauchst

df_results = df_fraud_aggregated_test_merged[["client_id"]].copy()
df_results["predicted_target"] = y_pred_test
df_results["predicted_proba"]  = y_proba_test

df_results.to_csv("fraud_baseline_test_predictions.csv", index=False)
```
